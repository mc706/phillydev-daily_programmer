````python
def create_game(name):
    def wrapped(guess):
        if guess.lower() == name.lower():
            return "CORRECT"
        if len(guess) != len(name):
            return "SHORT" if len(guess) < len(name) else "LONG"
        else:
            return len(set(guess.lower()) & set(name.lower()))
    return wrapped
````

````python
from itertools import permutations

def find_name(game):
    guesses = 0
    guess = ''
    stage = 'length'
    length = 1
    letters = []
    alpha = 'abcdefghijklmnopqrstuvwxyz'
    result = last_result = None
    sequences = []
    while True:
        if stage == 'length':
            guess = alpha[0:length]
        elif stage == 'letters':
            letter = guess[0]
            index = alpha.find(letter)
            if result == 1:
                letters.append(letter)
                if len(letters) == length:
                    stage = 'permutations'
            if index < 25:
                guess = alpha[index + 1] * length
            else:
                stage = 'permutations'
        elif stage == 'permutations':
            if not sequences:
                sequences = list(permutations(letters * (length - len(letters) + 1), length))
            guess = "".join(sequences.pop())
            
        guesses += 1
        result = game(guess)
        print('guessed {0}, got {1}, found {2}'.format(guess, result, letters))
        if result == "CORRECT":
            break
        elif result in ['SHORT', 'LONG']:
            stage = 'length'
            length = length + (1 if result == 'SHORT' else -1)
        elif stage != 'permutations':
            stage = 'letters'
    return guesses
````
